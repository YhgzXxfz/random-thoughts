```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Data
x = np.linspace(0, 10, num=11, endpoint=True)
y = np.cos(-(x**2) / 9.0)

# Interpolation
linear_interpolation = interp1d(x, y, kind="linear")
cubic_interpolation = interp1d(x, y, kind="cubic")

# Draw picture
x_finer_data = np.linspace(0, 10, num=41, endpoint=True)
plt.plot(x, y, "mo", label="data")  # [color][marker][line]
plt.plot(x_finer_data, linear_interpolation(x_finer_data), "c-", label="linear")
plt.plot(x_finer_data, cubic_interpolation(x_finer_data), "y--", label="cubic")

# Metadata
plt.legend() # plt.legend(["data", "linear", "cubic"], loc="best")
plt.title("Interpolation")
plt.xlabel("X", fontdict={"fontname": "Comic Sans MS"})
plt.show()
```

```python
# Data
labels = ['A', 'B', 'C']
values = [1, 4, 2]

# Chart
bars = plt.bar(labels, values)

# Apply patterns
patterns = ['/', 'o', '*']
_ = [bar.set_hatch(pattern) for (bar, pattern) in zip(bars, patterns)]

# Metadata
plt.title("Bar Chart")
plt.figure(figsize=(6, 4))
plt.show()
```
