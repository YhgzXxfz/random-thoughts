```python
import typing as tp
from collections import defaultdict


def solution(queries) -> tp.List[str]:
    courses_ = {}
    students_ = defaultdict(int)
    registration_ = defaultdict(list)
    grades_ = {}
    gpa_ = {}

    def create_course(course_id, course_name, credit):
        if course_id in courses_:
            return "false"

        for _, course_info in courses_.items():
            if course_name == course_info[0]:
                return "false"

        courses_[course_id] = (course_name, int(credit), "Standard")
        return "true"

    def register_for_course(student_id, course_id):
        if student_id not in students_:
            students_[student_id] = 24

        if course_id not in courses_:
            return ""

        credit = courses_[course_id][1]
        if students_[student_id] < credit:
            # not enough credit
            return ""

        if student_id in registration_[course_id]:
            # already registered
            return ""

        registration_[course_id].append(student_id)
        students_[student_id] -= credit
        return str(students_[student_id])

    def get_pairs():
        result = []
        for course_id, stud in sorted(registration_.items()):
            if course_id not in courses_:
                continue
            if courses_[course_id][2] != "Standard":
                continue

            candidates = sorted(stud)
            pairs = []
            for i in range(len(candidates)):
                for j in range(i + 1, len(candidates)):
                    pairs.append((candidates[i], candidates[j]))

            pairs = sorted(pairs, key=lambda p: (p[0], p[1]))
            result.append(", ".join(f"[{p[0]}, {p[1]}]" for p in pairs))
        return ", ".join(f"[{r}]" for r in result if r)

    def create_course_ext(course_id, name, credit, grading_type):
        if course_id in courses_:
            return "false"

        for _, course_info in courses_.items():
            if name == course_info[0]:
                return "false"

        courses_[course_id] = (name, int(credit), grading_type)
        return "true"

    def set_component_grade(student_id, course_id, component, grade):
        if student_id not in students_:
            return "invalid"
        if course_id not in courses_:
            return "invalid"
        if course_id not in registration_:
            return "invalid"
        if student_id not in registration_[course_id]:
            return "invalid"

        if student_id not in grades_:
            grades_[student_id] = {course_id: {component: int(grade)}}
            return "set"

        if course_id not in grades_[student_id]:
            grades_[student_id].update({course_id: {component: int(grade)}})
            return "set"

        if component not in grades_[student_id][course_id]:
            grades_[student_id][course_id].update({component: int(grade)})
            return "set"

        grades_[student_id][course_id][component] = int(grade)
        return "updated"

    def get_gpa(student_id):
        if student_id not in grades_:
            return ""

        for course_id, students in registration_.items():
            if student_id in students:
                if course_id not in grades_[student_id]:
                    return ""

                if len(grades_[student_id][course_id].keys()) < 3:
                    return ""

        total_standard_course = 0
        total_grades = 0
        num_pass = 0
        num_fail = 0
        for course_id, grade_info in grades_[student_id].items():
            if courses_[course_id][2] == "Standard":
                credit = courses_[course_id][1]
                total_standard_course += credit
                for _, grade in grade_info.items():
                    total_grades += grade * credit
            else:
                if sum(grade_info.values()) >= 66:
                    num_pass += 1
                else:
                    num_fail += 1

        gpa = int(total_grades / total_standard_course)
        gpa_[student_id] = (gpa, num_pass, num_fail)
        return ", ".join([str(i) for i in [gpa, num_pass, num_fail]])

    def find_nominee():
        for student_id in students_.keys():
            get_gpa(student_id)

        departments = list(sorted(set([course_id[:3] for course_id in courses_.keys()])))
        candidates = {dep: [] for dep in departments}
        for department in departments:
            stud_union = set()
            eligible_stud = set()
            for course_id in courses_.keys():
                if not course_id.startswith(department):
                    continue
                for stud in registration_[course_id]:
                    if stud in stud_union:
                        eligible_stud.add(stud)
                    stud_union.add(stud)
            candidates[department] = list(eligible_stud)

        nominee = {}
        for dep, studs in candidates.items():
            cands = sorted([stud for stud in studs if stud in gpa_])
            if not cands:
                nominee[dep] = ""
                continue

            best_grade = 0
            nom = cands[0]
            for cand in cands:
                if gpa_[cand][0] > best_grade:
                    best_grade = gpa_[cand][0]
                    nom = cand
            nominee[dep] = nom

        return ", ".join(f"{dep}({nom})" for dep, nom in nominee.items())

    result = []
    for q in queries:
        if q[0] == "CREATE_COURSE":
            result.append(create_course(q[1], q[2], q[3]))
        elif q[0] == "REGISTER_FOR_COURSE":
            result.append(register_for_course(q[1], q[2]))
        elif q[0] == "GET_PAIRED_STUDENTS":
            result.append(get_pairs())
        elif q[0] == "CREATE_COURSE_EXT":
            result.append(create_course_ext(q[1], q[2], q[3], q[4]))
        elif q[0] == "SET_COMPONENT_GRADE":
            result.append(set_component_grade(q[1], q[2], q[3], q[4]))
        elif q[0] == "GET_GPA":
            result.append(get_gpa(q[1]))
        elif q[0] == "FIND_NOMINEE":
            result.append(find_nominee())

        # print(courses_)
    return result


queries = [
    ["CREATE_COURSE_EXT", "CSE220", "Data Structures", "3", "Standard"],
    ["CREATE_COURSE_EXT", "CSE300", "Operating Systems", "4", "Standard"],
    ["CREATE_COURSE_EXT", "CSE330", "Computer Architecture", "3", "Standard"],
    ["CREATE_COURSE_EXT", "BIO102", "Microbe", "2", "Standard"],
    ["CREATE_COURSE_EXT", "BIO103", "Introduction to Biology", "2", "Standard"],
    ["CREATE_COURSE_EXT", "BIO205", "Computational Biology", "3", "Standard"],
    ["REGISTER_FOR_COURSE", "stu001", "CSE220"],
    ["SET_COMPONENT_GRADE", "stu001", "CSE220", "homeworks", "30"],
    ["SET_COMPONENT_GRADE", "stu001", "CSE220", "midterm", "23"],
    ["SET_COMPONENT_GRADE", "stu001", "CSE220", "final", "33"],
    ["REGISTER_FOR_COURSE", "stu001", "CSE330"],
    ["SET_COMPONENT_GRADE", "stu001", "CSE330", "homeworks", "25"],
    ["SET_COMPONENT_GRADE", "stu001", "CSE330", "midterm", "25"],
    ["SET_COMPONENT_GRADE", "stu001", "CSE330", "final", "25"],
    ["REGISTER_FOR_COURSE", "stu002", "CSE220"],
    ["SET_COMPONENT_GRADE", "stu002", "CSE220", "homeworks", "25"],
    ["SET_COMPONENT_GRADE", "stu002", "CSE220", "midterm", "23"],
    ["SET_COMPONENT_GRADE", "stu002", "CSE220", "final", "33"],
    ["REGISTER_FOR_COURSE", "stu002", "CSE330"],
    ["SET_COMPONENT_GRADE", "stu002", "CSE330", "homeworks", "25"],
    ["SET_COMPONENT_GRADE", "stu002", "CSE330", "midterm", "25"],
    ["SET_COMPONENT_GRADE", "stu002", "CSE330", "final", "15"],
    ["REGISTER_FOR_COURSE", "stu002", "CSE300"],
    ["SET_COMPONENT_GRADE", "stu002", "CSE300", "homeworks", "20"],
    ["SET_COMPONENT_GRADE", "stu002", "CSE300", "midterm", "20"],
    ["SET_COMPONENT_GRADE", "stu002", "CSE300", "final", "35"],
    # stu003 registered for 1 course
    ["REGISTER_FOR_COURSE", "stu003", "BIO102"],
    ["SET_COMPONENT_GRADE", "stu003", "BIO102", "homeworks", "30"],
    ["SET_COMPONENT_GRADE", "stu003", "BIO102", "midterm", "23"],
    ["SET_COMPONENT_GRADE", "stu003", "BIO102", "final", "33"],
    # stu004 registered for 2 courses, but has grade for only one of them
    ["REGISTER_FOR_COURSE", "stu004", "BIO102"],
    ["SET_COMPONENT_GRADE", "stu004", "BIO102", "homeworks", "30"],
    ["SET_COMPONENT_GRADE", "stu004", "BIO102", "midterm", "23"],
    ["SET_COMPONENT_GRADE", "stu004", "BIO102", "final", "33"],
    ["REGISTER_FOR_COURSE", "stu004", "BIO103"],
    ["SET_COMPONENT_GRADE", "stu004", "BIO103", "homeworks", "30"],
    ["SET_COMPONENT_GRADE", "stu004", "BIO103", "final", "33"],
    ["REGISTER_FOR_COURSE", "stu004", "CSE300"],
    ["SET_COMPONENT_GRADE", "stu004", "CSE300", "homeworks", "20"],
    ["SET_COMPONENT_GRADE", "stu004", "CSE300", "midterm", "20"],
    ["SET_COMPONENT_GRADE", "stu004", "CSE300", "final", "35"],
    # stu005 has 2 BIO courses graded.
    ["REGISTER_FOR_COURSE", "stu005", "BIO102"],
    ["SET_COMPONENT_GRADE", "stu005", "BIO102", "homeworks", "30"],
    ["SET_COMPONENT_GRADE", "stu005", "BIO102", "midterm", "23"],
    ["SET_COMPONENT_GRADE", "stu005", "BIO102", "final", "13"],
    ["REGISTER_FOR_COURSE", "stu005", "BIO205"],
    ["SET_COMPONENT_GRADE", "stu005", "BIO205", "homeworks", "30"],
    ["SET_COMPONENT_GRADE", "stu005", "BIO205", "midterm", "30"],
    ["SET_COMPONENT_GRADE", "stu005", "BIO205", "final", "13"],
    ["FIND_NOMINEE"],
]
result = solution(queries)
print(result)
```
